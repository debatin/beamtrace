```python
#A beam is defined as [[px,py,pz],[qx,qy,qz]] where p is the direction and q is the position

class surface:
    center=np.array([0,0,2.])
    R=1 
    n=1
    d=50
    def __init__(self,pos=1,power=0,n1=1.5,n2=1,diameter=50):
        self.R=(n2-n1)/(power+1.E-40)
        self.n=n2
        self.center=np.array([0,0,self.R+pos])
    def normal(self,points):   #return surface normals at points=[p1,p2,...,pn] where pi are points on surface 
            v=points-self.center
            return (v.T/np.linalg.norm(v,axis=1)).T*np.sign(self.R)
    def plot(self):
        R=self.R
        ylim=min(abs(R)*0.95,self.d/2)
        y=np.linspace(self.center[1]-ylim,self.center[1]+ylim,100)
        z=self.center[2]-np.sign(self.R)*np.sqrt(self.R**2-y**2)
        plt.plot(z,y)
        
    #a beam is defined as q the position and P the direction
class beam_field:
    Q=np.array([[0,0,0],[0,0.1,0]])   #position of the beams [q1,q2,...,qn] where qi is the 3d pos of each beam
    P=np.array([[0,0.1,1],[0,0.1,1]]) #direction of the beams [p1,p2,...,pn]
    path=np.array([Q])                #[Q1,Q2,...,QM] positions of the beam bundle where Qi=[q1,...,qn] is defined above 
    P_path=np.array([P])
    def rectangular_grid(self,nx,ny,d):   #create an array of parallel beams d durchmesser 
        if nx == 1:
            dx=0
        else:
            dx=d
        dy=d
        if ny==1:
            dy=0
        xv, yv = np.mgrid[-dx:dx:nx*1j,-dy:dy:ny*1j]
        self.Q=np.array([xv.flatten(),yv.flatten(),np.zeros(nx*ny)]).T        
        self.P=np.tile(np.array([0,0,1]),(nx*ny,1))

    def surface_points(self,surface):  #points where the beams hit the surface
        P=self.P
        Q=self.Q
        QC=surface.center-Q
        QCP=np.sum(QC*P,1)
        QC2=np.sum(QC*QC,1) 
        P2=np.sum(P*P,1)
        
        a=(QCP-np.sign(surface.R)*np.sqrt(QCP*QCP+P2*(surface.R**2-QC2)))/P2
        
        return (P.T*a).T+Q
    def refract(self,s,points):    #s surface at which the refraction takes place 
        n1=1                #for a normalized propagation vector P and a surface normal N and an outgoing P2
                            # the refraction laws can be written as n1*NxP=n2*NxP2
        P=(self.P.T/np.linalg.norm(self.P,axis=1)).T #normalized propagation vectors                    
        P=self.P #normalized propagation vectors                    
        n2=s.n
       
        N=s.normal(points)
        c=n1/n2*np.cross(N,P)                         #c= n1/n2 NxP,   equation c=NxP2 can be written as
        NC=np.cross(N,c)                              #P2=Nxc/N^2+b*N with b=sqrt(1-(NxC/N^2)^2) to normalize P2
        self.P=-n2*(NC+(np.sqrt(1-np.sum(NC*NC,1))*N.T).T)
        #print(self.P)
    def propagate(self,surfaces):
        self.path=np.array([self.Q])
        self.P_path=np.array([self.P])
        for s in surfaces:
            sp=np.array([self.surface_points(s)])
            
            self.path=np.append(self.path,sp,axis=0)
            self.refract(s,sp[0])    
            self.Q=sp[0]
            self.P_path=np.append(self.P_path,[self.P],axis=0)
    def plot(self):
        plt.plot(self.path[:,:,2],self.path[:,:,1],color='black',lw=0.1)
    def calculate_intersections(self,P1,Q1,P2,Q2):
        #print((P1[:,:,1]-P1[:,:,2]*P2[:,:,1]/P2[:,:,2]))
        x=(Q2[:,:,1]-Q1[:,:,1]-P2[:,:,1]/P2[:,:,2]*(Q2[:,:,2]-Q1[:,:,2]))/(P1[:,:,1]-P1[:,:,2]*P2[:,:,1]/P2[:,:,2])
        #print( Q1+x.T*P1,x)
        return Q1+x.T*P1,x
    def calculate_plane(self,z):
        P=self.P_path
        Q=self.path
        #print(((z-Q[-2,:,[2]])/P[-2,:,[2]]).T)
        #print(P[-2])
        return ((z-Q[-2,:,[2]])/P[-2,:,[2]]).T*P[-2]+Q[-2]
        
        
bb=beam_field()        
bb.rectangular_grid(1,20,0.9)
s1=surface()
s1.n=1.5
s2=surface()
s2.center=np.array([0,0,102])
s2.R=100
s3=surface()
s3.center=np.array([0,0,154])
s3.R=150
s1.normal(bb.P)
bb.propagate([s1,s2,s3])
bb.plot()
i,x=bb.calculate_intersections(bb.P_path[-2:-1,1:],bb.path[-2:-1,1:],bb.P_path[-2:-1,:-1],bb.path[-2:-1,:-1])
print(i)
plt.plot(i[0,:,2],i[0,:,1],color='black')
s1.plot()
s2.plot()
plt.show()
```

```python
from scipy.optimize import minimize
class lens_system:  
    """ Handle a whole system of lenses"""
    surfaces=[]
    """list containing the surfaces of the lens system"""
    entrance_pupil=11
    """diameter of entrance pupil"""
    marginal_ray=beam_field()
    """marginal_ray"""
    def add_surface(self,s):
        """add a surface to the system. Surfacess must be added ordered by their z value."""
        self.surfaces.append(s)
    def plot(self):
        for s in self.surfaces:
            s.plot()
    def calculate_marginal(self):
        """calculate path of marginal ray"""
        m=self.marginal_ray
        m.P=np.array([[0,0,1]])
        m.Q=np.array([[0,self.entrance_pupil,0]])
        m.propagate(self.surfaces)
        
        m.plot()
    def plot_einhuellende(self):
        f=beam_field()        
        f.rectangular_grid(1,20,self.entrance_pupil)
        f.propagate(self.surfaces)
        i,x=f.calculate_intersections(f.P_path[-2:-1,1:],f.path[-2:-1,1:],f.P_path[-2:-1,:-1],f.path[-2:-1,:-1])
        plt.plot(i[0,:,2],i[0,:,1],color='black')
    def calc_width_helper2(self,z):
        f=beam_field()        
        f.rectangular_grid(1,40,self.entrance_pupil)
        f.propagate(self.surfaces)
        
        pl=f.calculate_plane(z)
        res=max(pl[:,1])
        print(res)
        return res
    def calc_width(self):
        f=lambda x:self.calc_width_helper2(x)
        m=self.marginal_ray
        start=(m.path[-1,0,2]-m.path[-2,0,2])/2
        res=minimize(f,(start), method='Nelder-Mead')
        return res.final_simplex
    def calculate_width_helper(self,d):
        m=self.marginal_ray

        n=beam_field()
        n.P=np.array([[0,0,1]])
        n.Q=np.array([[0,d,0]])
        n.propagate(self.surfaces)
        i,x=m.calculate_intersections(m.P_path[[-2]],m.path[-2:-1],n.P_path[-2:-1],n.path[-2:-1])
        
        return x[0,0],i
    def calculate_width(self):
        f=lambda x:-self.calculate_width_helper(x)[0]
        res=minimize(f, (-self.entrance_pupil*0.8), method='Nelder-Mead')
        d=res.final_simplex[0][0][0]
        i=self.calculate_width_helper(d)[1]
        #plt.plot(i[0,:,2],i[0,:,1],'o',color='black')
        #print(i[0,0,1])
        return i

    
power=0.0110
p1=0.008201
power2=0.0086
p2=+0.00001
z1=130.8
power3=-0.00880

z0=15.1
s1=surface(z0,0.000001,1,1.5)
s2=surface(z0+5,power3,1.5,1)
s3=surface(z1,p2,1,1.5)
s4=surface(z1+7,power2-p2,1.5,1)
s5=surface(153,power,1,1.5)
s6=surface(160,0.000001,1.5,1)

s7=surface(240,0.0001,1,1.1)
ls=lens_system()
ls.add_surface(s1)
ls.add_surface(s2)
ls.add_surface(s3)
ls.add_surface(s4)
ls.add_surface(s5)
ls.add_surface(s6)
ls.add_surface(s7)
ls.plot()
ls.plot_einhuellende()
ls.calculate_marginal()
a=ls.calculate_width_helper(-0.1)
#print(a)
i=ls.calculate_width()
print(i)
res=ls.calc_width()
print(res)
#plt.xlim(137.0,140.4)
#plt.ylim(-0.025,0.003)
bb=beam_field()        
bb.rectangular_grid(1,50,11)
bb.propagate(ls.surfaces)
bb.plot()
plt.show()
```
